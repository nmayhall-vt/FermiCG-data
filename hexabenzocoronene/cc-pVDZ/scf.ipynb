```python
F = mf.get_fock()
C = mf.mo_coeff
S = mf.get_ovlp()

# Just use alpha orbitals
Cdocc = mf.mo_coeff[:,mf.mo_occ==2]
Cvirt = mf.mo_coeff[:,mf.mo_occ==0]
ndocc = Cdocc.shape[1]
nvirt = Cvirt.shape[1]

```

# Define Fragments by AOs

```python
# Find AO's corresponding to atoms
atom_frags = []
frag1=[0,1,2,3,40,41]
frag2=[6,7,8,9,10,11]
frag3=[12,13,14,15,16,17]
frag4=[20,21,22,23,24,25]
frag5=[26,27,28,29,30,31]
frag6=[34,35,36,37,38,39]
frag7=[4,5,18,19,32,33]
atom_frags.append(frag1)
atom_frags.append(frag2)
atom_frags.append(frag3)
atom_frags.append(frag4)
atom_frags.append(frag5)
atom_frags.append(frag6)
atom_frags.append(frag7)


nfrags = len(atom_frags)

full = []
frags = [[] for i in range(nfrags)]
for fidx in range(nfrags):
    print([i for i in atom_frags[fidx]])
    for ao_idx,ao in enumerate(mf.mol.ao_labels(fmt=False)):
        if ao[0] in  atom_frags[fidx]:
            if ao[2] in ("2pz"):
                if ao[3] in ("z"):
                    frags[fidx].append(ao_idx)
                    full.append(ao_idx)
print(frags)
```

# Construct initial full active space 
For each orbital block (Docc, Virt), project each subspace onto the full list of fragment AOs. This will determine our full active space.

```python
#
#   Get full active space
(Oact, Vact), (Oenv, Venv) = svd_subspace_partitioning_nonorth((Cdocc, Cvirt), full, S)


print(" Number of Active Doubly Occupied: %12.8f" % Oact.shape[1])
print(" Number of Active Doubly Virtual : %12.8f" % Vact.shape[1])
Cact = np.hstack((Oact, Vact))

pyscf.tools.molden.from_mo(mf.mol, "Cact.molden", Cact)
print(" Should be 1: ", np.linalg.det(Cact.T @ S @ Cact))
```

# Split active space into fragments

```python
# Project active orbitals onto fragments
init_fspace = []
clusters = []
Cfrags = []
orb_index = 1


for fi,f in enumerate(frags):
    print()
    print(" Fragment: ", f)
    (Of, Vf), (_, _) = svd_subspace_partitioning_nonorth((Oact, Vact), f, S)

    Cfrags.append(np.hstack((Of, Vf)))
    ndocc_f = Of.shape[1]
    init_fspace.append((ndocc_f, ndocc_f))
    nmof = Of.shape[1] + Vf.shape[1]
    clusters.append(list(range(orb_index, orb_index+nmof)))
    orb_index += nmof


# Orthogonalize Fragment orbitals
Cfrags = sym_ortho(Cfrags, S)
Cenv = 1.0 * Oenv

# Pseudo canonicalize fragments
C_act=canonicalize(Cfrags, F)
Cact = np.hstack(C_act)

print(np.linalg.det(Cact.T @ S @ Cact))
# Write Molden files for visualization
pyscf.tools.molden.from_mo(mf.mol, "Cact.molden", Cact)
assert( abs(np.linalg.det(Cact.T @ S @ Cact)) > 1e-12 )

print(" init_fspace = ", init_fspace)
print(" clusters    = ", clusters)

```

# Make Integrals

```python
print(Cenv.shape)
print(Cact.shape)
d1_embed = 2 * Cenv @ Cenv.T

h0 = pyscf.gto.mole.energy_nuc(mf.mol)
h  = pyscf.scf.hf.get_hcore(mf.mol)
j, k = pyscf.scf.hf.get_jk(mf.mol, d1_embed, hermi=1)

print(h.shape)
h0 += np.trace(d1_embed @ ( h + .5*j - .25*k))

h = Cact.T @ h @ Cact;
j = Cact.T @ j @ Cact;
k = Cact.T @ k @ Cact;
nact = h.shape[0]

h2 = pyscf.ao2mo.kernel(pymol, Cact, aosym="s4", compact=False)
h2.shape = (nact, nact, nact, nact)
# The use of d1_embed only really makes sense if it has zero electrons in the
# active space. Let's warn the user if that's not true

S = pymol.intor("int1e_ovlp_sph")
n_act = np.trace(S @ d1_embed @ S @ Cact @ Cact.T)
if abs(n_act) > 1e-8 == False:
    print(n_act)
    error(" I found embedded electrons in the active space?!")

h1 = h + j - .5*k;

np.save("ints_h0", h0)
np.save("ints_h1", h1)
np.save("ints_h2", h2)
np.save("mo_coeffs", Cact)
np.save("overlap_mat", S)
```
